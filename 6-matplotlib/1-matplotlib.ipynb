```python
#
x = np.linspace(0,4*np.pi) #0 - 12
y=np.sin(x) #sin(x)
plt.plot(x,y)
```

```python
y2 = np.cos(x)#plot
plt.plot(x,y)
plt.plot(x,y2)
```

```python
plt.plot(y,marker='o')
```

```python
# 
plt.figure(figsize=(12,3))
ax = plt.subplot(1,3,1)
ax.plot(x,y)
ax.grid(color = 'red')
```

```python
line = plt.plot(x,y)
print(line) #list
line[0].set_color('red') #
```

```python
angle = np.linspace(0,2*np.pi,200) #

x=np.sin(angle)
y=np.cos(angle)
plt.figure(figsize=(5,5)) #
plt.plot(x,y)
```

```python
plt.axis('off')
```

```python
plt.plot(x,y)
plt.xlim([-1.5,1.5])
plt.ylim([-1,1])
```
