```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
```

```python
img = plt.imread('./image1.jpg')
plt.imshow(img)
img.shape
```

```python
gray = img.mean(axis=-1) #-1就是最后一维   变二维,颜色维度没了
plt.imshow(gray,cmap=plt.cm.gray)  #编程黑白 由两个进行决定 
gray.shape
```

```python
gray = img.mean(axis = 2) #求平均值变黑白 有很多方式
plt.imshow(gray,cmap=plt.cm.gray)
gray.shape
```

```python
#只含有单一曲线的图
x = np.linspace(0,4*np.pi) #0 - 12
y=np.sin(x) #函数sin(x)
plt.plot(x,y)
```

```python
y2 = np.cos(x)#plot函数的两次调用
plt.plot(x,y)
plt.plot(x,y2)
```

```python
plt.plot(y,marker='o')
```

```python
#一行 三列
plt.figure(figsize=(12,3))
ax = plt.subplot(1,3,1)
ax.plot(x,y)
ax.grid(color = 'red')
```

```python
line = plt.plot(x,y)
print(line) #list
line[0].set_color('red') #设置这个线条为红色
```

```python
angle = np.linspace(0,2*np.pi,200) #园

x=np.sin(angle)
y=np.cos(angle)
plt.figure(figsize=(5,5)) #变圆
plt.plot(x,y)
```

```python
plt.axis('off')
```

```python
plt.plot(x,y)
plt.xlim([-1.5,1.5])
plt.ylim([-1,1])
```

```python
plt.plot(x,y)
plt.axis('equal')
plt.xlabel('X')
```
