```python
df = DataFrame({'item':np.random.randint(0,10,size=100),
                'sailer':np.random.randint(0,10,size=100),
                'weight':np.random.randint(30,300,size=100),
               'price':np.random.randint(1,20,size=100)})
df
```

```python
def conver_item(x):
    if x<2:
        return '萝卜'
    elif x<7:
        return '白菜'
    else:
        return '青椒'
df['item'] = df['item'].map(conver_item)
```

```python
df
```

```python
def convert_sailer(x):
    if x<5:
        return '张大妈'
    elif x<7:
        return '李大妈'
    else:
        return '赵大爷'
df['sailer'] = df['sailer'].map(convert_sailer)
df
```

```python
def convert_weight(x):
    if x<80:
        return 50
    elif x<150:
        return 100
    else:
        return 150
df['weight'] = df['weight'].transform(convert_weight)
df
```

```python
#统计这个月 张大妈 李大妈 周大爷销售多少
ret = df.groupby(['sailer'])['weight']
ret
```

```python
#分组的,但是还没有计算,销售就是一个求和
ret.sum()
```

```python
ret.min()
```

```python
ret.apply(np.sum)
```

```python
  ret.count()  #次数
```

```python
ret.apply(np.size)
```

```python
def count(x):
    return x.size
```

```python
ret.apply(count)
```

```python
def analysis(x):
    return (x.size,x.sum(),x.mean(),x.min())

ret.apply(analysis)
```
