```python
#数据清洗:
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
```

```python
df = DataFrame(np.random.randint(0,150,size=(150,3)),columns=['Python','Math','En'])
df
```

```python
for i in range(15):
    index = np.random.randint(0,150,size=1)[0]
    column = np.random.randint(0,3,size=1)[0]
    df.iloc[index,column] = None
```

```python
df
```

```python
df.info
```

```python
cond = df.isnull().any(axis = 1) #返回的还是一个dataframe
df[cond]
```

```python
#非空数据
cond = df.notnull().all(axis=1) #这一行都不为空
df[cond]
```

```python
df.dropna() #将null的删除  
```

```python
df.drop(labels=['En'],axis=1)
```

```python
cond=df.isnull().any(axis=1)
index = df[cond].index
index
```

```python
df2 = df.drop(labels=index)
df2
```

```python
df[cond]
```

```python
cond = (df2<60).any(axis=1)
index = df2[cond].index
df2.drop(labels=index)
```

```python
#平均分小于60的进行过滤
cond = df2.mean(axis=1) <90

index = df2[cond].index
df2.drop(index)
```

```python
#小于30分的,大于100分的,
cond1=df2.mean(axis=1)<60
cond1
```

```python
cond2 = df2.mean(axis=1)>100
cond2
```

```python
#或运算
cond = cond1 | cond2
df2[cond]
```

```python
df = DataFrame(np.random.randint(0,150,size=(5,3)),columns=['Python','En','Math'])
df
```

```python
df['Python'][0]=None
df['En'][3]=np.NAN
```

```python
df
```

```python
df.fillna(60) #填充一个固定值
```

```python
df.fillna(value=df.mean()) #填充平均分
```

```python
df.fillna(value=df.median()) #填充中位数
```

```python
df.fillna(method= 'pad',axis=1) #method : {'backfill', 'bfill', 'pad', 'ffill', None
```
