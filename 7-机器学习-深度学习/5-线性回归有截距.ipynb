```python
    #什么是梯度?  就是导数,一直导,诺必达   梯度下降是求解最大值min nax最快的速度
    #导数是函数变化率最快的方向 ,下山找坡度最抖的
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x:(x - 4.3)**2 + 7.5*x+6  #最小值 有 开口向上
```

```python
g = lambda x: 2*x -1.1  #让导数0   函数是0.55
```

```python
x = np.linspace(-5,9,100)
plt.plot(x,f(x)) #最小值在0-1之间
```

```python
result = []
#导数,一开始蒙一个数,后面再慢慢逼近
#设置起始值的大小
v_min = np.random.randint(-10,20,size=1)[0]
result.append(v_min)
#该值记录梯度下降下一步的值
v_min_last = v_min +1

#什么时候梯度下降结束,当精度达到
precision = 0.00001

#在一个退出条件  循环的次数
max_time=3000
#步幅:step
step = 0.01
count=0
print("--随机生成的最小值:",v_min)
while True:
    if np.abs(v_min - v_min_last)<precision:
        break
    if count>max_time:
        break
        
    v_min_last = v_min
    result.append(v_min)
    print('-----------梯度下降更新的最小值:',v_min)
    v_min = v_min-g(v_min)*step  #step就是在原来的基础上,减少的百分比
    count += 1
    
```

```python
plt.figure(figsize=(12,9))
x = np.linspace(-8,5,100)
plt.plot(x,f(x))

#绘制最小值的点
plt.scatter(result,f(np.array(result)))
```
