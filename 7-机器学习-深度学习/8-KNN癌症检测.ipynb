```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
cancer = pd.read_csv('./data_cancer/cancer2.csv')
cancer
```

```python
X = cancer.iloc[:,2:]
y=cancer.iloc[:,10:11]
y.head()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score #测试评估包
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)

y_ = knn.predict(X_test)
display(y_,y_test.values)
```

```python
accuracy_score(y_test,y_) #准确率0.97 这是我的想法
```

```python
knn.score(X_test,y_test) #0.9770992366412213
```

```python
#知道多少个没病的 成有病的,有病的成没病
pd.crosstab(index=y_test.values.T,columns= y_) #求表数据的差异
```
